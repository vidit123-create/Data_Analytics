```python
#Import libraries:-
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Fuel dataset:-
```

```python
df=pd.read_excel('C:\\Users\\arpit\\Downloads\\Fuel_Data_(Rev.01-25-Feb-20).xlsx')
df
```

```python
#Master dataset:-
```

```python
df1=pd.read_excel('C:\\Users\\arpit\\Downloads\\Master.xlsx')
df1
```

```python
#Merge dataset:-
```

```python
df2=pd.merge(df,df1,on='PTROCRD',how='inner')
df2
```

```python
#Comparision
```

```python
df2['mismatched_tank_capacity']=df2['AU_MOTOR']>df2['ORIGINAL_TANK_CAPACITY']
df2
```

```python
#Subsetting:-
```

```python
df3=df2[df2['mismatched_tank_capacity']==True]
df3
```

```python
#Groupby:-
```

```python
x=df3.groupby(['PTROCRD','MODEL','mismatched_tank_capacity']).agg({'mismatched_tank_capacity':'count','EXPAMT':'sum'})
x
```

```python
#Rename:-
```

```python
x.rename({'mismatched_tank_capacity':'Count_mismatched_tank_capacity'},axis=1,inplace=True)
x
```

```python
#Assign to a new column.
```

```python
df4=x
df4
```

```python
#Sort the dataset:-
```

```python
df5=df4.sort_values(by='EXPAMT',ascending=False)
df5
```

```python
#Write to a excel
```

```python
df5.to_excel('Fuel021.xlsx')
```

```python
#Take the top-10 data to show in the graph with three columns PTROCRD,EXPAMT and Count_mismatched_tank_capacity i.e.mismatched.
```

```python
df1=df[['PTROCRD','EXPAMT','Count_mismatched_tank_capacity']].head(10)
df1.reset_index(drop=True)
```

```python
#Graph:-
```

```python
fig,ax=plt.subplots(figsize=(15,6))
df1.plot.bar(x='PTROCRD',y='EXPAMT',width=0.2,color='violet',ax=ax,label='PTROCRD VS EXPAMT VS COUNT')
df1['Count_mismatched_tank_capacity'].plot(secondary_y=True,ax=ax)
ax.set_xticklabels(df1['PTROCRD'],rotation=90)
plt.title('PTROCRD VS EXPAMT & COUNT')
plt.legend()
plt.xlabel('PTROCRD')
plt.ylabel('AMOUNT VS COUNT')
plt.show()
```
